```python
transaction_per_day = df.groupby('Date').size()
```

```python
transaction_per_day.plot(kind='bar', figsize=(12, 6))
```

```python
weekly_transactions = (
df.set_index("Date")
.resample("W")
.size()
)
weekly_transactions.plot(kind="bar", figsize=(12, 6))
```

```python
df['weekday'] = df['Date'].dt.day_name()
transactions_per_weekday = df.groupby("weekday").size()
```

```python
transactions_per_weekday.plot(kind="bar", figsize=(12, 6))
```

```python
pivot = pd.pivot_table(
    df, 
    index='weekday',
    columns = 'Product line',
    aggfunc= 'size', 
    
)

top_product_lines = pivot.idxmax(axis = 1)
top_sales_per_weekday = pivot.max(axis = 1)

result = pd.DataFrame({
    'Top Product Line' : top_product_lines,
    'Sales Count': top_sales_per_weekday
})

result
```

```python
ax = result["Sales Count"].plot(kind="bar", figsize=(10,5))

for i, v in enumerate(result["Sales Count"]):
    ax.text(i, v + 0.5, result["Top Product Line"][i], ha='center')

plt.ylabel("Sales Count")
plt.title("Top Product Line per Weekday")
plt.show()
```

```python
gender_product = pd.pivot_table(
    df,
      index = "Product line", 
    columns = "Gender", 
    aggfunc = "size"
)

gender_product 
```

```python
gender_product.plot(kind="bar", figsize=(12, 6))

plt.xlabel("Product Line")
plt.ylabel("Count")
plt.title("Sales by Product Line and Gender")
plt.legend(title = "Gender")
plt.xticks(rotation=45)
plt.show()
```

```python
gender_total_price = pd.pivot_table(
    df,
      index = "Gender", 
    values = "Total", 
    aggfunc = "sum"
)

gender_total_price.plot(kind="bar", figsize=(12, 6))
```

```python
gender_total_price = pd.pivot_table(
    df,
      index = "Gender", 
    values = "Total", 
    columns = "Product line",
    aggfunc = "sum"
)

gender_total_price.plot(kind="bar", figsize=(12, 6))
plt.xlabel("Gender")
plt.ylabel("Total Sales")
plt.title("Total Sales by Gender and Product Line")
plt.legend(title = "Product line")
plt.xticks(rotation=0)
plt.show()
```

```python
city_product_count = pd.pivot_table(
 df,
  index="City",
    columns="Product line",
    aggfunc="size"
)
                                                    
mean_city_counts = city_product_count.mean(axis=1)

ax = city_product_count.plot(kind="bar", figsize=(12,6))
mean_city_counts.plot(
    ax=ax,
        marker="o",
    color="black",
    linewidth=2,
    label="Mean count"
)


plt.ylabel("Transaction Count")
plt.title("Transactions by City and Product Line")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

                                                
```

```python
sns.boxplot(data = df, x = "City", y = "Total")
plt.boxplot
plt.xlabel("City")  
plt.ylabel("Total Sales")
plt.title("Total distribution by city")
plt.xticks(rotation=0)
plt.show() 

# outliers 
# plt.boxplot()
```

```python
group_city_product_line = df.groupby(['City', 'Product line'])['Total'].sum().reset_index()

city_product_line_pivot = group_city_product_line.pivot(
    index='City',
      columns='Product line',
        values='Total')

mean_total = city_product_line_pivot.mean(axis=1)

ax = city_product_line_pivot.plot(kind='bar', figsize=(12, 6))
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.title("Total Sales by City and Product Line")
plt.legend(
    title="Product line",
    loc="lower right",
    bbox_to_anchor=(1.02, 1),  # move legend outside right
    borderaxespad=0
)
plt.tight_layout()

plt.xticks(rotation=0)
mean_total.plot(
    ax =ax,
    marker="o",
    color="black",
    linewidth=2,
    label="Mean Total"
)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# top_city_product_line = city_product_line_pivot.idxmax()
# values = city_product_line_pivot.max()

# top_city_product_line
# result = pd.DataFrame({
#     'Top Product Line' : top_city_product_line,
#     'Total Sales': values
# })

# result.plot(kind='bar', figsize=(12, 6))
# plt.xlabel("City")
# plt.ylabel("Total Sales")
# plt.title("Top Product Line by City")
# plt.legend(title = "Product line")
# plt.xticks(rotation=0)

# plt.show()
```
